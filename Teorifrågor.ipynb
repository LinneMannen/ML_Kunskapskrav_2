11. Vad är ”regressionsproblem? Kan du ge några exempel på modeller som används
och potentiella tillämpningsområden?

Regressionsproblem handlar om att med hjälp av modeller som exempelvis Linjär regression, lasso eller SVM prediktera et kontinuerligt värde. Kan användas till exmepelvis att prediktera bilpriser, huspriser eller löner.

12. Hur kan du tolka RMSE och vad används det till.

RMSE står för root mean squared error och är ett mätvärde för att utvärdera en modell. Den mäter precis som det låter roten av medelvärdet av alla fel i kvardrat och är ofta ett en bra utgångspunkt eftersom att du får resultatet i samma "enhet" som modellen predikterar. 

13. Vad är ”klassificieringsproblem? Kan du ge några exempel på modeller som
används och potentiella tillämpningsområden?

Det handlar om att med hjälp av modeller som exemeplvis Logistic regression, SVM och random forest prediktera kategorier eller klasser. Det kan användas till att exempelvis prediktera om någon är sjuk eller ej, spammfilter till en mail eller kanske prediktera vilken siffra (klass) som är handskriven på en bild.
 
14. Vad är en ”Confusion Matrix”?

Det är en tabell som visar varje klass och varje prediktering mot varje klass. Diagonalen i denna blir då de korrekta prediktionerna. Med hjälp av denna kan man sedan få ut olika mätvärden eller bara få en visuell översyn på hur modellen presterat.

15. Vad är K-means modellen för något? Ge ett exempel på vad det kan tillämpas på.

K-means är en unsupervised klustringsmodell som delar in data i K kluster igenom att tilldela punkter till närmsta centroid. Den kan användas för exemeplvis kundsegmentering.

18. Vad är skillnaden mellan parametrar och hyperparametrar i en
maskininlärningsmodell? Ge ett exempel på varje och förklara varför de inte kan
optimeras på samma sätt.

Parametrar = De värden som lärs av modellen exempelvis vikter och koefficienter
Hyperparametrar = Sätts före träning, ofta med hjälp av GridSearch och bestämmer egentligen hur modellen ska lära sig. Exempel på dessa är gamma, max_depth, epsilon och C.
De optimeras olika helt enkelt för att hyperparametrar väljs medans parametrar beräknas av modellen.

19. Förklara skillnaden mellan overfitting och underfitting i en
maskininlärningsmodell. Beskriv även hur man kan upptäcka respektive åtgärda
dem.

    Overfitting
Upptäcks av att du får bra resultat på train data och dåligt på test. Som åtgärd så kan man regularisera eller byta till en enklare modell

    Underfitting
Upptäcks av att du får dåligt resultat på både train och test. Detta sker oftast för att modellen inte hittar rätt mönster. Modellen är helt enkelt för regulariserad eller för enkel. Som åtgärd skulle man kunna testa en annan modell eller ändra med hyperparametrarna.
